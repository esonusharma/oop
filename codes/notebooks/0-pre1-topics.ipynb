# Topics

1. Construct a program that illustrates the concept and implementation of a class and Objects

Write a C++ program to calculate the density using class and object.

2. Create a program to showcase the use of friend function.

Write a C++ program to calculate the density showcasing friend class.
Write a C++ program to calculate the density showcasing friend function.

3. Create a program that implements constructor (simple, copy, parametrized) and destructor.

Write a C++ program to calculate the density using a default constructor.
Write a C++ program to calculate the density using a parameterized constructor.
Write a C++ program to calculate the density using a copy constructor.

4. Write a program that showcases the application of Encapsulation

Write a C++ program to calculate the weight density of water demonstrating encapsulation.

5. Construct a program that demonstrates implementation of Inheritance- Single and Multilevel

Write a C++ program to demonstrate single inheritance by calculating the density and relative density of a fluid.
Write a C++ program to demonstrate multilevel inheritance by calculating the density, relative density, and weight of a fluid.

6. Construct a program that demonstrates implementation of Inheritance- Multiple, Hierarchical and Hybrid

Write a C++ program to demonstrate multiple inheritance by calculating the density, relative density, and weight of a fluid.
Write a C++ program to demonstrate hierarchical inheritance by calculating the density, relative density, and weight of a fluid.
Write a C++ program to demonstrate hybrid inheritance by calculating the density, relative density, and weight of a fluid.

7. Create a program that exhibits the usage of Operator Overloading (Compile Time Polymorphism)

Write a C++ program to demonstrate operator overloading by implementing vector addition.

8. Create a program that exhibits the usage of Function Overloading (( Compile Time Polymorphism)

Write a C++ program to demonstrate function overloading by calculating areas of different shapes.

9. Construct a program that demonstrates the implementation of Abstract Classes and Virtual Functions( Run Time Polymorphism)

Write a C++ program to demonstrate abstract classes and virtual functions by calculating the density, relative density, and weight of a fluid.

10. Create a program to showcase various File Handling operations – open, close, update

Write a C++ program to demonstrate file handling operations - open, close, and update using density calculation.

11. Create a program to show the use of Command Line Arguments

Write a C++ program to demonstrate the use of Command Line Arguments for calculating stress.

12. Create a program to showcase the use of Templates.

Write a C++ program to demonstrate the use of Function Templates for calculating stress.

13. Create a program to implement Exception Handling – Try, Catch

Write a C++ program to demonstrate the use of Exception Handling for calculating stress.
